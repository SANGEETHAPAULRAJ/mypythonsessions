```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 8, 12, 6, 14]

# Create a line plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('product')
plt.ylabel('Y-axis label')
plt.title('Line Plot Example')

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
categories = ['Category A', 'Category B', 'Category C']
values = [30, 50, 25]

# Create a bar chart using Seaborn
sns.barplot(x=categories, y=values)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
data = [3, 5, 5, 6, 7, 8, 8, 9, 9, 9, 10, 11, 11, 12, 12, 12, 13, 14]

# Create a histogram using Matplotlib
plt.hist(data, bins=10, edgecolor='black')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')

# Display the plot
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data (correlation matrix)
data = np.random.rand(5, 5)

# Create a heatmap using Seaborn
sns.heatmap(data, annot=True, cmap='coolwarm')

# Add title
plt.title('Heatmap Example')

# Display the plot
plt.show()
```

```python


import matplotlib.pyplot as plt

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [30, 20, 15, 35]

# Create a pie chart using Matplotlib
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90)

# Add title
plt.title('Pie Chart Example')

# Display the plot
plt.show()
```

```python
import plotly.graph_objects as go

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 8, 12, 6, 14]
z = [5, 9, 3, 7, 2]

# Create a 3D scatter plot using Plotly
fig = go.Figure(data=go.Scatter3d(x=x, y=y, z=z, mode='markers'))

# Add title
fig.update_layout(title='3D Scatter Plot Example')

# Show the plot
fig.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
years = [2010, 2011, 2012, 2013, 2014]
series_a = [20, 25, 30, 35, 40]
series_b = [15, 18, 25, 28, 30]

# Create an area plot using Matplotlib
plt.stackplot(years, series_a, series_b, labels=['Series A', 'Series B'], alpha=0.6)

# Add labels and title
plt.xlabel('Years')
plt.ylabel('Values')
plt.title('Area Plot Example')
plt.legend(loc='upper left')

# Display the plot
plt.show()
```
