```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from statsmodels.graphics.regressionplots import influence_plot
import statsmodels.formula.api as smf
```

```python
cars=pd.read_csv("/content/sample_data/mtcars.csv")
```

```python
import statsmodels.formula.api as smf
model=smf.ols('mpg~cyl+disp+hp',data=cars).fit()
```

```python
model.params
```

```python
print(model.tvalues,'\n',model.pvalues)
```

```python
model.rsquared,model.rsquared_adj
```

```python
ml_c=smf.ols('mpg~cyl',data=cars).fit()
```

```python
print(ml_c.tvalues,'\n',ml_c.pvalues)
```

```python
rsq_hp=smf.ols('hp~cyl+mpg',data=cars).fit().rsquared
```

```python
vif_hp=1/(1-rsq_hp)
```

```python
import statsmodels.api as sm
qqplot=sm.qqplot(model.resid,line='q')
plt.title("Normal Q-Qplot of residuals")
plt.show()
```

```python
list(np.where(model.resid>10))
```

```python
def get_standarized_values(vals):
  return(vals-vals.mean())/vals.std()
  plt.scatter(get_standardized_values(model.fittedvalues), 
            get_standardized_values(model.resid))
  plt.title('residual plot')
  plt.xlabel('SFV')
  plt.ylabel('SRV')
  plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model,"hp",fig=fig)
plt.show()
```

```python
model_influence=model.get_influence()
(c,_)=model_influence.cooks_distance
```

```python
fig=plt.subplots(figsize=(20,7))
plt.stem(np.arange(len(cars)),np.round(c,3))
plt.xlabel('row')
plt.ylabel('cooks distance')
plt.show()
```

```python
(np.argmax(c),np.max(c))
```
