```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
#create DataFrame
df = pd.DataFrame({'points': [18, np.nan, 19, 14, 14, 11, 20, 28, 30, 31,
                              35, 33, 29, 25, 25, 27, 29, 30, 19, 23],
                   'assists': [3, 3, 4, 5, 4, 7, 8, 7, 6, 9, 12, 14,
                               np.nan, 9, 4, 3, 4, 12, 15, 11],
                   'rebounds': [15, 14, 14, 10, 8, 14, 13, 9, 5, 4,
                                11, 6, 5, 5, 3, 8, 12, 7, 6, 5]})

#view first five rows of DataFrame
print(df.head())
```

```python
#drop rows with NA values in any columns
df = df.dropna()

#create scaled DataFrame where each variable has mean of 0 and standard dev of 1
scaled_df = StandardScaler().fit_transform(df)

#view first five rows of scaled DataFrame
print(scaled_df[:5])
```

```python
#initialize kmeans parameters
kmeans_kwargs = {
"init": "random",
"n_init": 10,
"random_state": 1,
}

#create list to hold SSE values for each k
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(scaled_df)
    sse.append(kmeans.inertia_)

#visualize results
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

```python
#instantiate the k-means class, using optimal number of clusters
kmeans = KMeans(init="random", n_clusters=3, n_init=10, random_state=1)

#fit k-means algorithm to data
kmeans.fit(scaled_df)

#view cluster assignments for each observation
kmeans.labels_

```

```python
#append cluster assingments to original DataFrame
df['cluster'] = kmeans.labels_

#view updated DataFrame
print(df)
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of the plots
sns.set(style="whitegrid")

# Create scatter plot of 'points' vs. 'assists' colored by cluster
plt.figure(figsize=(10, 6))
sns.scatterplot(x='points', y='assists', hue='cluster', data=df, palette='Set1')
plt.title('Clustered Data: Points vs. Assists')
plt.xlabel('Points')
plt.ylabel('Assists')
plt.show()

# Create scatter plot of 'points' vs. 'rebounds' colored by cluster
plt.figure(figsize=(10, 6))
sns.scatterplot(x='points', y='rebounds', hue='cluster', data=df, palette='Set1')
plt.title('Clustered Data: Points vs. Rebounds')
plt.xlabel('Points')
plt.ylabel('Rebounds')
plt.show()

# Create scatter plot of 'assists' vs. 'rebounds' colored by cluster
plt.figure(figsize=(10, 6))
sns.scatterplot(x='assists', y='rebounds', hue='cluster', data=df, palette='Set1')
plt.title('Clustered Data: Assists vs. Rebounds')
plt.xlabel('Assists')
plt.ylabel('Rebounds')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of the plots
sns.set(style="whitegrid")

# Create violin plots for each variable within each cluster
for variable in ['points', 'assists', 'rebounds']:
    plt.figure(figsize=(10, 6))
    sns.violinplot(x='cluster', y=variable, data=df, palette='Set1')
    plt.title(f'Cluster-wise Distribution: {variable}')
    plt.xlabel('Cluster')
    plt.ylabel(variable)
    plt.show()
```
